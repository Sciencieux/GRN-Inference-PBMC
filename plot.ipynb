```python
import matplotlib.pyplot as plt

# 
methods = ['DIRECT-NET', 'SCODE', 'FigR', 'Regdiffusion']
pbmc3k_times = [31.2 * 60, 45.6 * 60, 27.8 * 60, 22.3 * 60]  # PBMC3K 
pbmc10k_times = [62.2 * 60, 91.3 * 60, 53.6 * 60, 40.7 * 60]  # PBMC10K 
pbmc3k_memory = [1500, 800, 1200, 900]  # PBMC3K 
pbmc10k_memory = [2500, 1200, 2000, 1500]  # PBMC10K 

# 
plt.figure(figsize=(10, 6))
bar_width = 0.3
index = range(len(methods))

# PBMC 3k
plt.bar(index, pbmc3k_times, bar_width, label='PBMC 3k', color='blue')
# PBMC 10k
plt.bar([i + bar_width for i in index], pbmc10k_times, bar_width, label='PBMC 10k', color='orange')

plt.xlabel('Methods')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime Comparison for Different Methods on PBMC Datasets (in seconds)')
plt.xticks(index, methods)
plt.legend()

plt.tight_layout()
plt.show()

# 
plt.figure(figsize=(10, 6))

# PBMC 3k
plt.bar(index, pbmc3k_memory, bar_width, label='PBMC 3k', color='blue')
# PBMC 10k
plt.bar([i + bar_width for i in index], pbmc10k_memory, bar_width, label='PBMC 10k', color='orange')

plt.xlabel('Methods')
plt.ylabel('Memory Usage (MB)')
plt.title('Memory Usage Comparison for Different Methods on PBMC Datasets')
plt.xticks(index, methods)
plt.legend()

plt.tight_layout()
plt.show()
```
