```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# data
methods = ['DIRECT-NET', 'SCODE', 'FigR', 'Regdiffusion']
memory_pbmc3k = [1500, 800, 1200, 900]
time_pbmc3k = [31.2, 45.6, 27.8, 22.3]
memory_pbmc10k = [2500, 1200, 2000, 1500]
time_pbmc10k = [62.2, 91.3, 53.6, 40.7]

colors = ['blue', 'green', 'red', 'cyan']

# Create axis and plot
fig, ax = plt.subplots()

# Linear fitting of data points for each algorithm
for i, method in enumerate(methods):
    # PBMC3K
    x3k = np.array([memory_pbmc3k[i], memory_pbmc10k[i]])
    y3k = np.array([time_pbmc3k[i], time_pbmc10k[i]])
    slope, intercept, r_value, p_value, std_err = linregress(x3k, y3k)
    
    # Draw fitting line
    ax.plot(x3k, y3k, label=f'{method} fit', color=colors[i])
    
    # Add data points and labels
    ax.scatter(x3k, y3k, color=colors[i], label=method)
    ax.text(memory_pbmc3k[i], time_pbmc3k[i], method, color=colors[i], fontsize=9)
    ax.text(memory_pbmc10k[i], time_pbmc10k[i], method, color=colors[i], fontsize=9)

# Set legend
ax.legend(title='Algorithms', loc='best')

# Set title and axis labels
ax.set_title('Algorithm Performance on Different Dataset Sizes')
ax.set_xlabel('Memory Usage (MB)')
ax.set_ylabel('Runtime (minutes)')

# Show the grid
ax.grid(True)

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Load dataa
methods = ['DIRECT-NET', 'SCODE', 'FigR', 'Regdiffusion']
memory_pbmc3k = [1500, 800, 1200, 900]  # PBMC3K memory usage
time_pbmc3k = [31.2 * 60, 45.6 * 60, 27.8 * 60, 22.3 * 60]  # PBMC3K run time (seconds)
memory_pbmc10k = [2500, 1200, 2000, 1500]  # PBMC10K memory usage
time_pbmc10k = [62.2 * 60, 91.3 * 60, 53.6 * 60, 40.7 * 60]  # PBMC10K run time (seconds)

# set the colour of each methods
colors = ['blue', 'green', 'red', 'cyan']

# Draw the fitting line
plt.figure(figsize=(12, 8))

for i, method in enumerate(methods):
    # Plot the scatter plot of the PBMC3K dataset
    plt.scatter(memory_pbmc3k[i], time_pbmc3k[i], color=colors[i], label=f'{method} ', zorder=5)
    # Plot the scatter plot of the PBMC10K dataset
    plt.scatter(memory_pbmc10k[i], time_pbmc10k[i], color=colors[i], zorder=5)
    
    # Draw the fitting line
    plt.plot([memory_pbmc3k[i], memory_pbmc10k[i]], [time_pbmc3k[i], time_pbmc10k[i]], 
             color=colors[i], linestyle='-', linewidth=1, label=f'{method} fit')

plt.title('Algorithm Performance on PBMC Datasets')
plt.xlabel('Memory Usage (MB)')
plt.ylabel('Runtime (seconds)')
plt.legend(title='Algorithms', loc='best')  # Adapt legend location
plt.grid(True)
plt.show()
```
